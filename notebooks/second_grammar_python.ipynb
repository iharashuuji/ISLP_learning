## Sequence and Slice Notation

```python
# linspace can make create an array. And, 
seq1 = np.linspace(0, 10, 11)
print(f"seq1 array: {seq1}, seq1 dtypes: {seq1.dtype}")
seq2 = np.arange(0, 10)
print(f"seq2 array: {seq2}, seq2 dtypes: {seq2.dtype}")
```

```python
# under code results in print text 'lo', this method start number is 0, and point 3 and 6. So,print text lo 
print("hello world"[3:6])
print("hello world"[slice(3, 6)])
```

```python
A = np.array(np.arange(16)).reshape((4, 4))
print('original vector')
print(A)
print('-'*40)
print(A[1 : 2]) # You think result is 6.but, print text is [4, 5, 6, 7]. this method.
print('-'*40)
print(f"A[1, 2] results: {A[1, 2]}\n\n A[:, [0, 2]] results \n {A[:, [0, 2]]}")
print('-'*40)
# under two code results in the same results.
print(A[[1, 3], [0, 2]]) # the results was caused by the method which is [[x, y], [x', y']] x and x' means [x, x']points num. 
print(np.array([A[1, 0], A[3, 2]]))
```

```python
# A[[1, 3], [0, 2, 3]] # this code un
print(A[[1, 3]][:, [0, 2]])
idx = np.ix_([1, 3], [0, 2, 3])
print(A[idx])
```

```python
A[1:4:2, 0:3:2]
print(f'A.shapee[0]: {A.shape[0]}') # this is A.shape = (4, 4), A.shape[0] accesses to first number, so A.shape[0] is 4.
keep_rows = np.zeros(A.shape[0], bool)
print(f"original keep_rows: {keep_rows}")
keep_rows[[1, 3]] = True
print(f"changed_keep_rows: {keep_rows}") # maybe changed second and forth elements.
```

```python
# If 
print(f"np.all(keep_rows == np.array([0, 1, 0, 1])): {np.all(keep_rows == np.array([0, 1, 0, 1]))}") # arrays elements True conditions has can display, other elments False conditions have can't display.
print(f"\nA[np.array([0, 1, 0, 1])]:\n {A[np.array([0, 1, 0, 1])]}") # first and second rows in original A display. So, [x, y, z, g] x elements display.
```

```python
print(f"keep_rows: {keep_rows}")
A[keep_rows] # display rows True conditions have in keep_rows. So, second, forth rows display.
```

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
print(f"keep_rows: {keep_rows}\nkeep_cols: {keep_cols} ")
idx_bool = np.ix_(keep_rows, keep_cols)
print(f"\nA[idx_bool]: \n{A[idx_bool]}")
```
