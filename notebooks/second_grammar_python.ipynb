```python
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
np.mean(y), y.mean()
```

```python
# Note: can calculate mean, std
# np.var means variance(分散)
# variance: $$V = E[X^2] - (E[X])^2$$
print(f"{np.var(y), y.var(), np.mean((y - y.mean()) ** 2)}")
print(f"sqrt: {np.sqrt(np.var(y))}, std: {np.std(y)}")
```

```python
# 10*3 matrix of N(0, 1)
# N(x, y) means that x is means, y is std, N is 
X = rng.standard_normal((10, 3))
X
```

```python
# Note: under two codes results in the same num.
print(X.mean(axis=0))
print(X.mean(0))
```
